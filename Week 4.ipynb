# Sales Data Analysis

```python
import pandas as pd

data = {
    "date": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04"],
    "product": ["A", "B", "A", "C"],
    "sales": [100, 200, 150, 300]
}

df = pd.DataFrame(data)
df.to_csv("sales_data.csv", index=False)
```

```python
df = pd.read_csv("sales_data.csv")
df
```

```python
df["date"] = pd.to_datetime(df["date"])
df.info()
```

```python
df.isnull().sum()
```

## Analysis Questions
1. Total sales by product

```python
total_sales = df.groupby("product")["sales"].sum()
total_sales
```

## Analysis Questions
2. Average daily sales

```python
avg_sales = df.groupby("product")["sales"].mean()
avg_sales
```

```python
total_sales.sort_values(ascending=False)
```

```python
summary = df.groupby("product")["sales"].agg(["sum", "mean"])
summary
```

## Analysis Results
- Produt C has the highest total sales.
- Product A shows stable average sales.
- Product B has fewer records but relatively high sales per day.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
summary["sum"].plot(kind="bar")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
summary["mean"].plot(kind="bar")
plt.title("Average Sales by Product")
plt.xlabel("Product")
plt.ylabel("Average Sales")
plt.show()
```

```python
daily_sales = df.groupby("date")["sales"].sum()
daily_sales
```

```python
daily_sales.plot()
plt.title("Daily Total Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

## Visualization Insights
- Product C shows the highest total salesoverall.
- Product A maintains relatively stable sales.
- Daily sales show an increasing trend toward the end of the period.

# Final Summary

- Product C recorded the highest total sales.
- Product A showed stable average sales.
- Daily sales increased toward the end of the period.
